모듈 없이 짜보는 선형회귀

```
x_data = [1,2,3,4]
y_data = [15,25,35,45]
```

```
def mul(listA, intB):
  output = []
  for i in listA:
    output.append(i*intB)
  return output

def mul2(listA, listB):
  output = []
  for i in range(len(listA)):
    output.append(listA[i]*listB[i])
  return output

def add(listA, intB):
  output = []
  for i in listA:
    output.append(i+intB)
  return output

def sub(listA, listB):
  output = []
  for i in range(len(listA)):
    output.append(listA[i]-listB[i])
  return output

def squre(listA):
  output = []
  for i in listA:
    output.append(i*i)
  return output

def mean(listA):
  output = 0
  for i in listA:
    output += i
  output /= len(listA)
  return output
```

```
W = 1 # 임의의 값
B = 1 # 임의의 값
def update(X,Y):
  global W
  global B
  hypo = add(mul(X, W), B)
  loss = mean(squre(sub(hypo, Y)))
  W = W - 0.1*mean(mul2(sub(add(mul(X,W),B),Y),X))
  B = B - 0.1*mean(sub(add(mul(X,W),B),Y))
  print(hypo,' | ',loss)
```

```
#test
for i in range(100):
  update(x_data,y_data)
```
