```
import tensorflow as tf
```

```
x_data = [[1,1],
          [2,2],
          [3,3],
          [4,4]]
y_data = [[0],
          [0],
          [1],
          [1]]
```

```
tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.Dense(1,activation='sigmoid',input_shape=(2,)))

tf.model.compile(optimizer = tf.keras.optimizers.SGD(0.1),
                 loss='mse',
                 metrics=['accuracy'])
tf.model.summary()
tf.model.fit(x_data,y_data,epochs=500)
```
